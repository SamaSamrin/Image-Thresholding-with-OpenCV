```python
import numpy as np
import cv2
print("OpenCV-Python Version {}".format(cv2.__version__))
```

```python
from matplotlib import pyplot as plt
```

### Taking the image as input

```python
image = cv2.imread("grayscale.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
assert image is not None, "file could not be read, check with os.path.exists()"
```

###Thresholding

The method returns two outputs. The first is the threshold that was used and the second output is the thresholded image.

```python
thresh = 127
max_val = 255
#src = source pixel value, dest = destination pixel value

#binary thresholding - if src > thresh, dest = max_val. Else dest = 0
ret, binary_thresh = cv2.threshold(image, thresh, max_val, cv2.THRESH_BINARY)

#inverse of binary thresholding
ret, binary_inverse_thresh = cv2.threshold(image, thresh, max_val, cv2.THRESH_BINARY_INV)

# truncate thresholding - If src > thresh, dest = thresh. Else, dest= src
ret, truncate_thresh = cv2.threshold(image, thresh, max_val, cv2.THRESH_TRUNC)

#to zero - if src > thresh, dest = src. Else, dest = 0
ret, zero_thresh = cv2.threshold(image, thresh, max_val, cv2.THRESH_TOZERO)

#inverse of to zero
ret, zero_inverse_thresh = cv2.threshold(image, thresh, max_val, cv2.THRESH_TOZERO_INV)
```

Plotting the images

```python
titles = ['Original Image', 'Binary', "Binary Inverse", "Truncate", "To Zero", "To Zero Inverse"]
images = [image, binary_thresh, binary_inverse_thresh, truncate_thresh, zero_thresh, zero_thresh]

for i in range(6):
  plt.subplot(2, 3, i+1), plt.imshow(images[i],'gray',vmin=0, vmax=255)
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
